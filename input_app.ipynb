```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import tensorflow as tf

# Load the saved model
loaded_model = tf.keras.models.load_model('kmer_model.h5')
loaded_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Create the main window with a larger size
root = tk.Tk()
root.title("Popup Window App")
root.geometry("500x300")  # Adjust the dimensions

def show_inputs():
    input1 = entry1.get()
    input2 = float(entry2.get())  # Convert to float
    input3 = float(entry3.get())  # Convert to float
    
    # Create an input DataFrame
    input_data = pd.DataFrame({
        'Kmer': [input1],
        'skew': [input2],
        'ratio': [input3]
    })
    
    # Preprocess the input data
    encoder = joblib.load('encoder.pkl')
    X_cat_test = encoder.transform(input_data[['Kmer']])
    X_cat_test = X_cat_test.reshape(1, X_cat_test.shape[1])  # Reshape to match the original shape



    numerical_cols = ['skew', 'ratio']
    scaler_X = joblib.load('standard_scaler.pkl')
    X_num_test = scaler_X.transform(input_data[numerical_cols])
    X_num_test = X_num_test.reshape(1, X_num_test.shape[1])
    
    predictions = loaded_model.predict([X_cat_test, X_num_test])
    
    # Show the predictions in a messagebox
    messagebox.showinfo("Predictions", f"Predicted C0 Value: {predictions}")

def reset_inputs():
    entry1.delete(0, tk.END)
    entry2.delete(0, tk.END)
    entry3.delete(0, tk.END)

# Create labels and entry fields for inputs
label1 = tk.Label(root, text="Kmer:")
entry1 = tk.Entry(root, width=30)

label2 = tk.Label(root, text="Skew:")
entry2 = tk.Entry(root, width=30)

label3 = tk.Label(root, text="Ratio:")
entry3 = tk.Entry(root, width=30)
sequence_length = 6
# Create buttons
show_button = tk.Button(root, text="Show Predictions", command=show_inputs)
reset_button = tk.Button(root, text="Reset Inputs", command=reset_inputs)

# Pack the widgets onto the window
label1.pack()
entry1.pack()
label2.pack()
entry2.pack()
label3.pack()
entry3.pack()
show_button.pack()
reset_button.pack()

# Start the main event loop
root.mainloop()
```
